## With size changing

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style='whitegrid')
tip= sns.load_dataset('tips')
plt.figure(figsize=(8,8))

sns.boxplot(x='day', y='tip', data= tip, hue='size',palette={"yellow", 'pink','black', 'red',
                                                             'green', 'purple'})
```

```python
import seaborn as sns
import pandas as pd
import numpy as np

kashti= sns.load_dataset('titanic')
kashti.head()
```

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

kashti= sns.load_dataset('titanic')
sns.boxplot(x= 'survived', 
            y='age', showmeans= True,
            meanprops= {'marker': '*', 'markersize': '12', 'markeredgecolor': 'white'},
            data=kashti)
#show lables
plt.xlabel("how many survived", size=14)
plt.ylabel("Their age", size=13)
plt.title("LABELED BOXPLOT", size=15, weight= 'bold')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style='whitegrid')
tip= sns.load_dataset("tips")
g= sns.FacetGrid(tip,row='size',col='day', sharex= False)
g.map(plt.scatter, 'tip', 'sex')
```
